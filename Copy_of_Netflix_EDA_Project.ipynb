# Netflix Exploratory Data Analysis (EDA) Project

This project analyzes the Netflix Movies and TV Shows dataset using Python.

Objective:
To clean the dataset, perform exploratory data analysis, create visualizations, identify trends, and generate meaningful business insights.

Tools Used:
- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn
- Google Colab

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!unzip netflix_titles.csv.zip
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["director"].fillna("Unknown", inplace=True)
```

```python
df["cast"].fillna("Unknown", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df["country"].fillna("Unknown", inplace=True)
```

```python
df.shape
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
df.isnull().sum()
```

```python
df["type"].value_counts()
```

```python
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

## Insight 1

Netflix has significantly more Movies than TV Shows.

The dataset shows that movies are the dominant type of content on Netflix, indicating that Netflix focuses more on movie-based entertainment.

## Insight 2

The United States contributes the highest number of titles on Netflix.

This indicates that Netflix has a strong content base from the US market, while countries like India and the United Kingdom also contribute significantly to the platform’s content library.

```python
df["rating"].value_counts()
```

```python
df["rating"].value_counts().head(10).plot(kind="bar")
plt.title("Top Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

## Insight 3

TV-MA is one of the most common content ratings on Netflix.

This suggests that a large portion of Netflix content is designed for mature audiences, showing a strong focus on adult entertainment and serious content categories.

```python
df["release_year"].value_counts().head(10)
```

```python
df["release_year"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Release Years on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

## Insight 4

Netflix released the highest number of titles in recent years.

This shows that Netflix has rapidly expanded its content library over time, especially after 2015, indicating strong growth in content production and global expansion.

```python
df["listed_in"].value_counts().head(10)
```

```python
df["listed_in"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Netflix Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.show()
```

## Insight 5

Drama and International Movies are among the most common content categories on Netflix.

This shows that Netflix focuses on diverse global entertainment and offers a wide variety of content to attract audiences from different countries and interests.

```python
df["type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Percentage of Movies and TV Shows")
plt.ylabel("")
plt.show()
```

# Final Overall Insights

1. Netflix has significantly more Movies than TV Shows.

2. The United States contributes the highest number of titles on Netflix.

3. TV-MA is one of the most common ratings, showing a strong focus on mature audience content.

4. Netflix released the highest number of titles in recent years, indicating rapid content growth.

5. Drama and International Movies are among the most common categories, showing Netflix focuses on global entertainment.

# Most Surprising Finding

The most surprising finding was that Netflix has a significantly higher number of Movies compared to TV Shows.

I expected TV Shows to be more common because streaming platforms are known for series-based entertainment. However, the dataset shows that Movies are the dominant content type on Netflix, which was unexpected.
